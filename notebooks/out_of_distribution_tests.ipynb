```python
import json 
import random

source_file = "../data/autojailbreaks/iris_gcg_suffixes_source_Llama3_converted_output.json"
with open(source_file, "r") as f:
    data = json.load(f)
    
print(len(data))
print(len( [d for d in data if d['toxic']]))  # Count toxic examples
print(len( [d for d in data if d['jailbreak_label']]))  # Count non-toxic examples

# test = random.sample([d for d in data if d['toxic']], k=1)
# for t in test:
#     print("Input:")
#     print(t['prompt'])
#     print("Output:")
#     print(t['output'])
#     print("Toxicity:", t['toxic'])
#     print("-----")


source_toxicity = {d['source'][0]: [] for d in data}
for d in data:
    source = d['source'][0]
    source_toxicity[source].append(d['toxic'])

# Calculate the average toxicity for each source
average_toxicity = {source: sum(toxicity) / len(toxicity) for source, toxicity in source_toxicity.items() if len(toxicity) > 0}

toxic_sources = [k for k in average_toxicity.keys() if average_toxicity[k] > 0.5]
print( f"Total sources: {len(source_toxicity)}")
print( f"Total sources with toxic examples: {len(toxic_sources)}")

universal_jailbreak_data = [d for d in data if average_toxicity.get(d['source'][0], 0) > 0.8]

toxic_ratio_per_source = {s: [d['toxic_ratio'] for d in universal_jailbreak_data if d['source'][0] == s and d['toxic']] for s in {d['source'][0] for d in universal_jailbreak_data}} 

for key, values in toxic_ratio_per_source.items():
    print( f"Source: {key}, Toxic Ratio: {sum(values) / len(values) if values else 0:.2f}, Toxic Ratio (outer): {average_toxicity.get(key, 0):.2f}" )    
    
# test = [d['jailbreak'] for d in data if d['jailbreak_label']]
# print("\n\n+++\n+++\n\n".join(test))

for key in toxic_sources:
    print(f"Source: {key}, Average Toxicity: {average_toxicity[key]:.2f}")
    prompt_scores = sorted([d for d in data if d['source'][0] == key and d['toxic']], key=lambda x: x['toxic_ratio'], reverse=True)
    for p in prompt_scores:
        print(f"{p['toxic_ratio']}: \n\t{p['prompt']}\n\t{p['jailbreak']}")

#detected = [d['prompt'] for d in data if d['jailbreak_label'] and d['source'][0] in toxic_sources]
detected = [d['prompt'] for d in data]

print(len(detected))
with open(source_file.replace("_converted_output.json", "_sample.json"), "w", encoding="utf-8") as f:
    json.dump(detected, f, indent=2, ensure_ascii=False)
```
